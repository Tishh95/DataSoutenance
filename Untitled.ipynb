```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

lol = pd.read_csv('League of Legends 2021 World Championship Play-In Groups Statistics - Raw Data.csv')

sup = lol[lol['Position'] == 'Support' ]

sup['Player'].value_counts()
```

```python
indexNames = sup[((sup['Ward Interactions'] > sup['Ward Interactions'].mean()) & (sup['Player'].value_counts()> 1))].index
sup.drop(indexNames , inplace=True)

plt.figure(figsize=(15,15))
sns.catplot(y="Kill Participation", x="Player", data=sup , kind="bar" , aspect=4);
```

```python
sns.catplot(y="Wards Placed", x="Player", data=sup , kind="bar" , aspect=4);
```

```python
top = lol[lol['Position'] == 'Top' ]
indexNames = top[(top['Kills'] > top['Kills'].mean())].index
top.drop(indexNames , inplace=True)
```

```python
plt.figure(figsize=(15,15))
sns.catplot(y="Kills", x="Player", data=sup , kind="bar" , aspect=4);
```
